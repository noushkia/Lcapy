# RL Circuit

```python
N = R(10) + L(0.01)

v_f = np.logspace(0, 5, 400)
Z_s = N.Z(f).evaluate(v_f)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.loglog(v_f, abs(Z_s), linewidth=2)
ax.set_ylabel('Impedance (ohms)')
ax.set_xlabel('Frequency (Hz)')
plt.show()
```
